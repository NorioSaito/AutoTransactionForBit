```python
import poloniex
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
%matplotlib inline

polo = poloniex.Poloniex()
polo.timeout = 2
rawdata = polo.returnChartData('USDT_BTC',
                                period = 300,
                                start=time.time()-polo.DAY*180,
                                end=time.time())

price_data = pd.DataFrame([float(i.get('open')) for i in rawdata])

plt.figure()
price_data.plot(figsize=(200,100), title='BTC_rawdata')
plt.savefig('rawdata.png')
```
